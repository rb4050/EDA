### find out avg price of laptop by brand 

```python
a = df.groupby(by="brand") \
    .agg({"Price":"mean"}) \
    .sort_values(by="Price")
```

```python
a
```

```python
a.plot(kind="barh")
```

### find out median price by rating and warranty

```python
b = df.groupby(by=["rating","warranty"]) \
    .agg({"Price":"median"}) \
    .sort_values(by="Price")
```

```python
b
```

```python
b.plot(kind="barh")
```
